### Guessing an elephant's weight
Remember when Alex mentioned an activity guessing an elephant's weight a while back? Well now, we have enough skills to complete that challenge!
You, along with 49 other people, have been asked to guess the weight of an adult male elephant. The table below shows all the guesses. Using this data set, answer the questions that follow.
Note: For this problem and the questions that follow, to avoid any ambiguity you should use the sample standard deviation formula rather than the population standard deviation formula. Furthermore, you should use only the sample standard deviation formula without the Bessel Correction (the n-1 in the denominator).  Some calculators will use this correction since sample statistics tend to underestimate the population standard deviation. In this way we can be consistent with the answers given here.


```python
import seaborn as sns
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
print(median)
```

```python
mean = df['Guess (lbs)'].mean()

print(mean)
```

```python
mode = df['Guess (lbs)'].mode()

print(mode)
```

```python
std = df['Guess (lbs)'].std()

print(std)
```

```python
percentile = df['Guess (lbs)'].quantile(0.46)

print(percentile)
```

```python
from scipy import stats
```

```python
z = np.abs(stats.zscore(df))
```

```python
outliers = df[z > 3]
```

```python
print(outliers)
```

```python
small_outliers = outliers[outliers < 0.1]

print(small_outliers)
```
